# Bar Chart — Product Category wise Sales

```python
import matplotlib.pyplot as plt
sample=df.groupby("Product_Category")["Sales_Amount"].mean()
plt.bar(sample.index, sample.values)
plt.xlabel("Product Category")
plt.ylabel("Total Sales Amount")
plt.title("Category-wise Sales")
plt.xticks(rotation=45)
plt.show()
```

# Line Graph — Date wise Sales Trend

```python
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

daily_sales = df.groupby("Sale_Date")["Sales_Amount"].sum()

plt.plot(daily_sales.index, daily_sales.values,color="black")
plt.xlabel("Date")
plt.ylabel("Sales Amount")
plt.title("Daily Sales Trend")
plt.show()
```

# Pie Chart — Payment Method Contribution

```python
sample=df.groupby("Payment_Method")["Sales_Amount"].mean()
plt.pie(sample.values,labels=sample.index)
plt.title("Payment Method Contribution")
plt.show()
```
