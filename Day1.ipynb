```python
import json
import os
file_name = "finance_data.json"
def load_data():
    os.path.exists(file_name)
    with open(file_name,"r") as file:
        return json.load(file)
def save_data(data):
    with open(file_name,"w") as file:
         json.dump(data,file,indent=4)
```

Adding transaction


```python
data=[]
def add_transc(data):
    t_type = input("Enter type (income/expense): ").lower()
    amount = float(input("Enter amount: "))
    category = input("Enter category: ").lower()
    month = input("Enter month (YYYY-MM): ")
    transaction = {
        "type": t_type,
        "amount": amount,
        "category": category,
        "month": month}
    data.append(transaction)   
    save_data(data)
    print("Success")

    
```

Monthly records

```python
def monthly_report(data):
    month = input("Enter month (YYYY-MM): ")
    income = 0
    expense = 0
    for t in data:
        if month == t["month"]:
            if t["type"] == "income":
                income += t["amount"]
            elif t["type"] == "expense":
                expense += t["amount"]
    print("Total Income:", income)
    print("Total Expense:", expense)
    print("Savings:", income - expense)
```
