Manual Dot Product


```python
def sample(a,b):
    result=0
    for i in range(len(a)):
        result+=a[i]*b[i]
    return result
```

```python
print(sample([1,2,3], [4,5,6]))
```

Vector Operations

```python
def vector_operations(a, b):
    print("Vector A:", a)
    print("Vector B:", b)
    print("Addition:", a + b)
    print("Subtraction:", a - b)
    print("Scalar Multiplication (2 * A):", 2 * a)
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
vector_operations(v1, v2)
```

# Matrix Multiplication (Manual)

```python
import numpy as np

def manual_matrix_multiplication(A, B):
    rows_A = len(A)
    cols_B = len(B[0])
    cols_A = len(A[0])

    # Step 1: empty result matrix (sab 0)
    result = []

    for i in range(rows_A):
        row = []
        for j in range(cols_B):
            sum_val = 0
            for k in range(cols_A):
                sum_val += A[i][k] * B[k][j]
            row.append(sum_val)
        result.append(row)

    return np.array(result)
A=np.array([[1,2],
            [2,3]])
b=np.array([[9,2],
            [7,3]])
manual_matrix_multiplication(A,b)
```
