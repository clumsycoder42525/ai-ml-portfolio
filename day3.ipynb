```python
import json
import os

FILE = "contacts.json"

class ContactBook:
    def __init__(self):
        if os.path.exists(FILE):
            with open(FILE, "r") as f:
                self.contacts = json.load(f)
        else:
            self.contacts = []

    def save(self):
        with open(FILE, "w") as f:
            json.dump(self.contacts, f, indent=4)
    contacts={}

    def add_contact(self):
        name = input("Name: ")
        phone = input("Phone: ")
        email = input("Email: ")

        self.contacts.append({
            "name": name,
            "phone": phone,
            "email": email
        })

        self.save()
        print("Contact added")

    def view_contacts(self):
        if not self.contacts:
            print("No contacts")
            return

        for i, c in enumerate(self.contacts, 1):
            print(i, c["name"], c["phone"], c["email"])

    def search_contact(self):
        name = input("Search name: ").lower()

        for c in self.contacts:
            if c["name"].lower() == name:
                print(c)
                return

        print("Contact not found")

    def edit_contact(self):
        name = input("Edit name: ").lower()

        for c in self.contacts:
            if c["name"].lower() == name:
                c["phone"] = input("New phone: ")
                c["email"] = input("New email: ")
                self.save()
                print("Contact updated")
                return

        print("Contact not found")

    def delete_contact(self):
        name = input("Delete name: ").lower()

        for c in self.contacts:
            if c["name"].lower() == name:
                self.contacts.remove(c)
                self.save()
                print("Contact deleted")
                return

        print("Contact not found")
```

```python
book = ContactBook()

while True:
    print("\n1 Add")
    print("2 View")
    print("3 Search")
    print("4 Edit")
    print("5 Delete")
    print("6 Exit")

    ch = input("Choice: ")

    if ch == "1":
        book.add_contact()
    elif ch == "2":
        book.view_contacts()
    elif ch == "3":
        book.search_contact()
    elif ch == "4":
        book.edit_contact()
    elif ch == "5":
        book.delete_contact()
    elif ch == "6":
        break
    else:
        print("Invalid choice")
```
