```python
import csv

class StudentRecords:
    def __init__(self):
        self.students = {}   
        
    def add_student(self, roll, name, marks):
        self.students[roll] = {
            "name": name,
            "marks": marks,
            "grade": self.calculate_grade(marks)
        }
        print("Student added successfully")

    def calculate_grade(self, marks):
        if marks >= 80:
            return "A"
        elif marks >= 60:
            return "B"
        else:
            return "C"

    def update_student(self, roll, marks):
        if roll in self.students:
            self.students[roll]["marks"] = marks
            self.students[roll]["grade"] = self.calculate_grade(marks)
            print("Student updated")
        else:
            print("Student not found")

    def delete_student(self, roll):
        if roll in self.students:
            del self.students[roll]
            print("Student deleted")
        else:
            print("Student not found")

    # Export to CSV
    def export_csv(self):
        with open("result.csv", "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["Roll", "Name", "Marks", "Grade"])

            for roll, s in self.students.items():
                writer.writerow([roll, s["name"], s["marks"], s["grade"]])

        print("Data exported")
```

```python
s1 = StudentRecords()

s1.add_student(1, "Rahul", 85)
s1.add_student(2, "Anita", 92)

s1.update_student(1, 88)
s1.delete_student(2)

```
